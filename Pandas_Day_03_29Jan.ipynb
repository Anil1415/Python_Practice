```python
import pandas as pd
```

```python
d1 = {'key1':[1,2,3,4,5,"sudh"],
      'key2' : [7,8,9,10,11,12],
      'key3':[13,14,15,16,17,18],
      'key4':[19,20,21,22,23,24]
}
```

```python
pd.DataFrame(d1)
```

```python
# dic to Dataframe (key as column name and values as column values)
df1 = pd.DataFrame(d1)
```

```python
df1
```

```python
# column name changes
pd.DataFrame(d1, index = ['a','b','c','d','e','f'])
```

```python
d2 = {'key1':[1,2,3,4,5,"sudh"],
      'key2' : [7,8,"kumar",10,11,12],
      'key3':[13,14,15,16,17,18],
      'key4':[19,20,21,"ineuron","pwskills",24]
}
```

```python
pd.DataFrame(d2)
```

```python
df2 = pd.DataFrame(d2)
```

```python
df2
```

```python
df1
```

```python
pd.merge(df2,df1)
```

```python
# df1 left join and df2 right join like venn diagram 'A' intersection 'B'
# missing rows with string values for inner merge operation

# 'A' intersecn 'B' == 'B' intersection 'A'

pd.merge(df1,df2)
```

```python
# on--> which column check and matching on 'key1'

# how = 'inner' is default.

pd.merge(df1,df2, how='inner', on='key1')
```

```python

pd.merge(df1,df2, how='inner', on='key2')
```

```python
pd.merge(df1,df2, how='inner', on='key4')
```

```python
# returns NaN if no matching data in the rows on the right side of 'key2'
pd.merge(df1,df2, how='left', on='key4')
```

```python
pd.merge(df1,df2, how='right', on='key4')
```

```python
pd.merge(df1,df2, how='outer', on='key4')
```

```python
 pd.merge(df1,df2, how='outer', on=['key4','key1'])
```
